LINE CHART

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data
data = {'Date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
        'Value': [10, 15, 20, 25, 30, 28, 32, 34, 40, 42]}
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Value'], marker='o')
plt.title('Line Chart')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()

```

This code creates a simple line chart using Python libraries pandas and matplotlib. It first imports matplotlib.pyplot for plotting and pandas for handling data. A dictionary is then defined containing two keys: Date, which uses pandas date_range to generate ten consecutive daily dates starting from January 1, 2023, and Value, which stores a list of numerical values. This dictionary is converted into a pandas DataFrame for easier data manipulation. The plotting section initializes a figure with a specified size, then plots the Date column on the x-axis and the Value column on the y-axis, adding circular markers to each data point. It sets a title for the chart and labels both axes for clarity, enables a grid to make the chart easier to read, and finally displays the line chart on the screen.


BAR CHART

A figure is created with a specified size to control the overall dimensions of the chart. Finally, the chart is given a title, the x-axis and y-axis are labeled to describe the data being shown, and the bar chart is displayed on the screen.

BUBBLE CHART

```python
import matplotlib.pyplot as plt

# Sample Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
sizes = [100, 200, 300, 400, 500]

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=sizes, alpha=0.5, color='green')
plt.title('Bubble Chart')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()

```

This code produces a bubble chart using the matplotlib library. It begins by importing matplotlib.pyplot for plotting. Three lists are defined: one for x-axis values, one for y-axis values, and one for sizes, which determine the area of each bubble. A figure is created with a specified size to control the display dimensions. The scatter function is used to plot the data points, where the s parameter assigns different bubble sizes, the alpha parameter makes the bubbles partially transparent, and the color parameter sets them to green. The chart is then given a title, both axes are labeled to describe the data, and finally the bubble chart is displayed on the screen.

AREA PLOT AND STACKED PLOT

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data
data = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'A': [10, 15, 20, 25, 30, 28, 32, 34, 40, 42],
    'B': [5, 7, 8, 6, 7, 5, 8, 9, 10, 11]
})

# Area Plot
plt.figure(figsize=(10, 6))
plt.fill_between(data['Date'], data['A'], color="skyblue", alpha=0.4, label='A')
plt.fill_between(data['Date'], data['B'], color="salmon", alpha=0.4, label='B')
plt.title('Area Plot')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

This code generates an area plot using matplotlib and pandas to visualize two sets of values over time. It begins by importing matplotlib.pyplot for plotting and pandas for data manipulation. A pandas DataFrame is created containing a Date column with ten consecutive daily dates starting from January 1, 2023, along with two numerical columns labeled A and B that represent different data series. A figure is initialized with a specified size to define the plot dimensions. The fill_between function is used twice to create shaded areas under the curves for series A and B across the date range, with different colors, partial transparency, and labels to distinguish them. The plot is then given a title, the x-axis and y-axis are labeled for clarity, a legend is added to identify each area, and finally the area plot is displayed on the screen.


PIE CHART

```python
import matplotlib.pyplot as plt

# Sample Data
labels = ['Category A', 'Category B', 'Category C']
sizes = [40, 30, 30]

# Plot
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Pie Chart')
plt.show()

```

This code creates a pie chart using the matplotlib library. It starts by importing matplotlib.pyplot, which provides the functions needed for plotting. The code then defines two lists: one containing category labels and another containing numerical values that represent the proportion of each category. A figure is created with equal width and height to ensure the pie chart appears circular. The pie function is used to draw the chart, where the sizes list determines the slice proportions, the labels list names each slice, the autopct parameter displays percentage values on the chart, and the colors parameter assigns custom colors to each slice. The chart is given a title for context, and finally the pie chart is displayed on the screen.


TABLE CHART

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data
data = pd.DataFrame({
    'Category': ['A', 'B', 'C'],
    'Value': [40, 30, 30]
})

# Plot
fig, ax = plt.subplots(figsize=(6, 2))
ax.axis('tight')
ax.axis('off')
ax.table(cellText=data.values, colLabels=data.columns, cellLoc='center', loc='center')
plt.title('Table Chart')
plt.show()

```

This code displays a simple table using matplotlib and pandas instead of a traditional chart. It begins by importing matplotlib.pyplot for visualization and pandas for data handling. A pandas DataFrame is created containing category labels and their corresponding values. The code then creates a figure and an axes object with a specified size, and the axes are adjusted by turning off the axis lines and tightening the layout so the table fits neatly. The table function is used to render the DataFrame values as a table within the plot area, with column headers displayed and all cell contents centered. A title is added above the table to describe it, and finally the table is displayed on the screen.


POLAR CHART

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
labels=np.array(['A', 'B', 'C', 'D'])
values=np.array([4, 3, 2, 5])
angles=np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()

# Complete the loop
values=np.concatenate((values,[values[0]]))
angles+=angles[:1]

# Plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='cyan', alpha=0.5)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
plt.title('Polar Chart')
plt.show()

```

This code creates a polar chart, also known as a radar or spider chart, using matplotlib and numpy. It begins by importing matplotlib.pyplot for plotting and numpy for numerical operations. The code defines arrays for category labels and their corresponding values, then calculates evenly spaced angles around a circle based on the number of labels. To close the shape of the radar chart, the first value is appended to the end of the values array and the first angle is added again to the angles list, ensuring the plot forms a complete loop. A figure and polar axes are created to enable circular plotting. The fill function is used to draw and shade the polygon representing the values, with partial transparency for better visibility. The radial tick labels are removed for a cleaner look, the angular ticks are set to match the category labels, a title is added, and finally the polar chart is displayed on the screen.


HISTOGRAM

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
data = np.random.randn(1000)

# Plot
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

This code generates a histogram using matplotlib and numpy to visualize the distribution of numerical data. It starts by importing matplotlib.pyplot for plotting and numpy for numerical operations. A sample dataset of one thousand random values drawn from a standard normal distribution is created using numpy. A figure is then initialized with a specified size to control the appearance of the plot. The hist function is used to divide the data into thirty bins and display the frequency of values in each bin, with purple bars and black edges to make the bars clearly distinguishable. The plot is given a title, the x-axis is labeled to represent the data values, the y-axis is labeled to show frequency, and finally the histogram is displayed on the screen.


LOLLIPOP CHART

```python
import matplotlib.pyplot as plt

# Sample Data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Plot
plt.figure(figsize=(10, 6))
plt.stem(categories, values, basefmt=" ")
plt.title('Lollipop Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

This code creates a lollipop chart using matplotlib to display values for different categories. It begins by importing matplotlib.pyplot, which provides plotting functions. The code defines a list of category labels and a corresponding list of numerical values. A figure is created with a specified size to control the charts dimensions. The stem function is used to draw vertical lines from a baseline to each data point, with markers at the top, creating a lollipop-style appearance, and the basefmt parameter is set to remove the baseline for a cleaner look. The chart is then given a title, the x-axis and y-axis are labeled to describe the data, and finally the lollipop chart is displayed on the screen.
